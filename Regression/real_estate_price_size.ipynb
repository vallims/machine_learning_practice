```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("data/real_estate_price_size.csv")
data.describe()
```

```python
x1 = data['price']
y = data['size']
```

```python
plt.scatter(x1,y)
plt.xlabel('price',fontsize = 20)
plt.ylabel('size',fontsize = 20)
plt.show()
```

```python
#Regression
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = -122.3349 + 0.0033*x1
fig = plt.plot(x1,yhat,c='green',lw=4,label='regression line')
plt.xlabel('price', fontsize = 20)
plt.ylabel('size', fontsize = 20)
plt.show()
```
